## Import Bechdel Data

```python
import pandas as pd 
import requests

# import bechdel data
response = requests.get('http://bechdeltest.com/api/v1/getAllMovies?')
json_array  = response.json()
data_bechdel = pd.DataFrame(json_array)
data_bechdel = data_bechdel[ data_bechdel['imdbid'] != '']
```

## The next step is to import the IMDB data

I downloaded the data earlier. ((https://datasets.imdbws.com/)

```python
# import imdb data title.basics.tsv.gz
data_imdb_basics = pd.read_csv('title.basics.tsv.gz', delimiter="\t", low_memory=False)
data_imdb_basics.columns = ['imdbid', 'titleType', 'primaryTitle', 'originalTitle', 'isAdult',
       'startYear', 'endYear', 'runtimeMinutes', 'genres']
data_imdb_basics['imdbid'] = data_imdb_basics['imdbid'].str.replace('tt', '')

# import imdb data title.ratings.tsv.gz
data_imdb_ratings = pd.read_csv('title.ratings.tsv.gz', delimiter="\t", low_memory=False)
data_imdb_ratings.columns = ['imdbid', 'averageRating', 'numVotes']
data_imdb_ratings['imdbid'] = data_imdb_ratings['imdbid'].str.replace('tt', '')
```

Next step is to join the data from the API and the IMDB data. 

```python
# join bechdel data, title.basics.tsv.gz, title.ratings.tsv.gz
data = data_bechdel.merge(data_imdb_basics, how = 'left', on = 'imdbid')
data = data.merge(data_imdb_ratings, how = 'left', on = 'imdbid')
data = data[['year', 'rating', 'primaryTitle', 'imdbid',  'isAdult', 'runtimeMinutes', 'genres'
             , 'averageRating', 'numVotes', 'titleType']]
data = data[data['titleType'] == 'movie'].reset_index(drop = True)
```

Encode the multilable attribute "genres" into 0 and 1, split comma-seperated strings into distinct strings

```python
# split into distinct strings
data['genres'] = data['genres'].fillna('')  
data['genres_labels'] = data['genres'].str.split(',')
```

```python
data['genres'].head()
```

```python
data['genres_labels'] = data['genres'].str.split(',').apply(lambda lst: [g.strip() for g in lst])
all_genres = pd.Series([g for sublist in data['genres_labels'] for g in sublist]).value_counts()
print(all_genres)
```

transform them into 0 and 1 

```python
# create the dummy variables
dummies = pd.get_dummies(data['genres_labels'].apply(pd.Series).stack(), prefix='')
dummies = dummies.groupby(level=0).sum()
```

```python
# create the dummy variables
dummies = pd.get_dummies(data['genres_labels'].apply(pd.Series).stack(), prefix='')
dummies = dummies.groupby(level=0).sum()
#dummies.drop(columns=columns_to_drop, axis=1, inplace=True)
df_combined = pd.concat([data['rating'] , dummies], axis=1)
```

```python
df_combined
```

```python
# calculate Bechdel test scores per movie genre
df_genre_ratings = df_combined.groupby('rating').sum().T
total_genres = df_genre_ratings.sum(axis=1)
df_genre_percentages = df_genre_ratings.div(total_genres, axis=0) * 100
df_genre_percentages = df_genre_percentages.reset_index()
df_genre_percentages = df_genre_percentages.sort_values(by=3, ascending=True)
```
